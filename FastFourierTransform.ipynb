```python
#In this notebook we will see how to do a simple fast fourier transform function
```

```python
import numpy as np #we need the numpy numerical module 
import matplotlib.pyplot as plt #Also we will need to plot later on, so let's import matplotlib too
```

```python
def fft(data):
    out = []
    N = len(data)
    if N==1: #If the data has just one element return the data, which is the fourier transform of data
        return data
    else:
        data_even = [] #We will store the even position data in here
        data_odd = [] #And the odd position data in here
        for i in range(N):
            if i%2:
                data_odd.append(data[i])
            else:
                data_even.append(data[i])
        fft_even = fft(data_even) #Lets calculate the fast fourier transform of the even data
        fft_odd = fft(data_odd) #Lets calculate the fast fourier transform of the odd data
        #And now let's calculate the fourier transform of data
        for i in range(N):
            theta = complex(0,2*np.pi*i/N)
            out.append(fft_even[i%(N//2)]+ fft_odd[i%(N//2)]*np.exp(theta))
        return out
```

```python
#Let's try our code
t = np.linspace(0,5,256)
x = np.sin(40*t)+5*np.cos(20*t)
#Let's compare our code with the fft function inside numpy module
plt.figure(1)
fft_x = fft(x)
plt.plot(np.abs(fft_x))

plt.figure(2)
y = np.fft.fft(x)
plt.plot(np.abs(y))
```
