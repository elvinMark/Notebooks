```python
#We include the modules that we are going to need for this program
import numpy as np
import matplotlib.pyplot as plt
```

```python
def runge_kutta_ode(fun_df,x0,t0,tf,N):
    u = [] #We initialize the output
    u.append(x0) #Push the first point
    time = np.linspace(t0,tf,N) #Generate the mesh of the time
    for i in range(1,len(time)):
        h = time[i] - time[i-1] #Calculate the time difference of the mesh
        k1 = h*fun_df(u[i-1],time[i-1])
        k2 = h*fun_df(u[i-1] + k1/2,time[i-1]+h/2)
        k3 = h*fun_df(u[i-1] + k2/2,time[i-1]+h/2)
        k4 = h*fun_df(u[i-1] + k3,time[i-1])
        u.append(u[i-1] + (1/6)*(k1 + 2*k2 + 2*k3 + k4)) #Calculate the next point using the Runge-Kutta Method
    return time,np.array(u)
```

```python
#Defining the difference function to use in Runge-Kutta method as an example
def fundf(u,t):
    out = np.zeros(2)
    out[0] = u[1]
    out[1] = -5*u[0]
    return out
```

```python
x0 = np.array([0,5])
t,u = runge_kutta_ode(fundf,x0,0,5,100)
plt.plot(t,u[:,0])
```
